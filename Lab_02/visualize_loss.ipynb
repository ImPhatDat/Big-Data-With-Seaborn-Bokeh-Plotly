```python
import matplotlib.pyplot as plt

# Read the content of the file
with open("saved_2_3/task_2_3.loss", "r") as file:
    lines = file.readlines()

# Extract iteration numbers and loss values
iterations = []
loss_values = []
for i in range(0, len(lines), 2):
    iterations.append(int(lines[i].strip().split("_")[1]))
    loss_values.append(float(lines[i+1].strip()))

# Combine iteration numbers and loss values into tuples
data = list(zip(iterations, loss_values))

# Sort based on iteration numbers
sorted_data = sorted(data, key=lambda x: x[0])

# Separate sorted iteration numbers and loss values
sorted_iterations = [item[0] for item in sorted_data]
sorted_loss_values = [item[1] for item in sorted_data]


# Plot the data
plt.plot(sorted_iterations, sorted_loss_values, marker='o')
plt.title('K-Means on preprocessed data')
plt.xlabel('Iteration')
plt.ylabel('WCSS loss')
plt.grid(True)
plt.show()
```
