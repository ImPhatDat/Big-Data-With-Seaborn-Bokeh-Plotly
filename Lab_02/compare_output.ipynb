```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Kết quả mong đợi

```python
df = pd.read_csv('2DPoints.csv')
sns.scatterplot(df, x='x1', y='x2', hue='class')
    
plt.xlim((0, 2))
plt.ylim((0, 2))

plt.show()
```
